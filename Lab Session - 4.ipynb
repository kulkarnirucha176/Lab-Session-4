### 1. Web Scraping of Job Qualifications using RAKE Algorithm

Parsing HTML data using BeautifulSoup library

```python
import urllib.request as ur
from bs4 import BeautifulSoup
job_url = ur.urlopen('https://lensa.com/business-analyst-business-systems-jobs/palo-alto/jd/f6e8dc5c747f056bb6433508df5779ca').read()
soup = BeautifulSoup(job_url, 'html.parser')
tableresult = soup.findAll("li")
records = []
for row in tableresult:
    records.append(row.text)
```

### 2. Using RAKE algorithm extracting the important keywords from Job Qualifications

RAKE refers to Rapid Automatic Keyword Extraction

```python
from rake_nltk import Rake
import nltk

r = Rake()
rankedRecords = []
for record in records :
    r.extract_keywords_from_text(record)
    rankedRecords.append(r.get_ranked_phrases())
```

### 3. Creating a dataframe with keywords and exporting it to CSV

```python
import pandas as pd
csv_data2 = pd.DataFrame()
for rankedRecord in rankedRecords :
    csv_data = pd.DataFrame({'Qualifications': rankedRecord})
    csv_data2 = csv_data2.append(csv_data)
```

```python
qualifications_table = csv_data2.copy()
qualifications_table.to_csv('QualificationsKeywords.csv', index=False)
```
